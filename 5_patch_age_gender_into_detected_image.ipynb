```python
import os
import numpy as np
import cv2
import matplotlib.pyplot as plt


frames_dir = './output/frames/'
gender_dir = './output/gender/'
ages_dir = './output/ages/'
dump_dir = './output/final/'

if not os.path.exists(dump_dir):
    os.makedirs(dump_dir)

names_list = [ele for ele in os.listdir(frames_dir) if ele.endswith('.jpg')]
names_list = sorted(names_list, key=lambda x: int(x.split('.jpg')[0]))

for name in names_list:
    print('Processing Frame {}'.format(name))
    merge_dump_path = os.path.join(dump_dir, name)
    
    detected_frame_path = os.path.join(frames_dir, name)
    ages_frame_path = os.path.join(ages_dir, name)
    gender_frame_path = os.path.join(gender_dir, name)
    
    background = cv2.imread(detected_frame_path, cv2.IMREAD_COLOR)
        
    if (not os.path.exists(ages_frame_path)) or (not os.path.exists(ages_frame_path)):
        cv2.imwrite(merge_dump_path, background)
    else:
        ages_in_top_layer = cv2.imread(ages_frame_path, cv2.IMREAD_COLOR)
        gender_in_top_layer = cv2.imread(gender_frame_path, cv2.IMREAD_COLOR)
        ages_in_top_layer = cv2.resize(ages_in_top_layer, (901, 543), interpolation=cv2.INTER_CUBIC)
        gender_in_top_layer = cv2.resize(gender_in_top_layer, (901, 543), interpolation=cv2.INTER_CUBIC)
        
        gender = np.zeros(shape=background.shape, dtype=np.uint8)
        gender[1903:2446, 3447:4348, :] = gender_in_top_layer
        
        age = np.zeros(shape=background.shape, dtype=np.uint8)
        age[1903:2446, 2529:3430, :] = ages_in_top_layer
        
        overlapping = cv2.addWeighted(background, 1.0, gender, 0.85, 0)
        overlapping = cv2.addWeighted(overlapping, 1.0, age, 0.85, 0)
        cv2.imwrite(merge_dump_path, overlapping)

    
    
```
