```python
import os
import cv2

class FacialRecog(object):
    def __init__(self, input_video_path, dump_frames_dir):
        self._input_video_path = input_video_path
        self._dump_frames_dir = dump_frames_dir

        if not os.path.exists(self._dump_frames_dir):
            os.makedirs(self._dump_frames_dir)

    def frames_dump(self):
        video = cv2.VideoCapture(self._input_video_path)
        frame_cnt = 0
        while True:
            res, image = video.read()
            if not res:
                break

            frame_dump_full_path = os.path.join(self._dump_frames_dir, '{}.jpg'.format(frame_cnt))
            cv2.imwrite(frame_dump_full_path, image)
            print('Writing frame {} into {}'.format(frame_cnt, frame_dump_full_path))
            frame_cnt += 1


if __name__ == '__main__':
    operator = FacialRecog(
        input_video_path='./Prime-Minister-of-Sweden-Visits-Volvo-Group-Trucks.mp4',
        dump_frames_dir='./raw_volvo_demo_video_frames/',
    )

    operator.frames_dump()
```
