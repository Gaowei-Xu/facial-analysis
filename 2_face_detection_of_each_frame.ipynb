```python
import boto3 
import json
import os

frames_root_dir = './raw_volvo_demo_video_frames/'
results_root_dir = './face_detection_results/'

if not os.path.exists(results_root_dir):
    os.makedirs(results_root_dir)

frames = [ele for ele in os.listdir(frames_root_dir) if ele.endswith('.jpg')]
frames = sorted(frames, key=lambda x:int(x.split('.')[0]))

client=boto3.client('rekognition')

for index, frame_name in enumerate(frames):
    full_path = os.path.join(frames_root_dir, frame_name)
    print('Processing frame {}'.format(frame_name))
    
    with open(full_path, 'rb') as image:
        image_bytes = image.read()        
        response = client.detect_faces(Image={'Bytes': image_bytes},Attributes=['ALL'])
        
        dump_json_path = os.path.join(results_root_dir, frame_name.split('.')[0] + '.json')
        json.dump(response, open(dump_json_path, 'w'), indent=True, ensure_ascii=False)
    
```
