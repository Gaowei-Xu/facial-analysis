```python
frames_root_dir = './output/final/'
output_video_full_path = './output/demo.mp4'

import cv2
import os

frames = [ele for ele in os.listdir(frames_root_dir) if ele.endswith('.jpg')]
frames = sorted(frames, key=lambda x: int(x.split('.jpg')[0]))

image = cv2.imread(os.path.join(frames_root_dir, frames[0]))
print(image.shape)

video_frame_size = (image.shape[1], image.shape[0])
fps = 30

fourcc = cv2.VideoWriter_fourcc('X', 'V', 'I', 'D')
video_writer = cv2.VideoWriter(output_video_full_path, fourcc, fps, video_frame_size)

for frame_name in frames:
    data = cv2.imread(os.path.join(frames_root_dir, frame_name), cv2.IMREAD_COLOR)
    video_writer.write(data)
    print('Processing {}'.format(frame_name))
    
video_writer.release()
```
