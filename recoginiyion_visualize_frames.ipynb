```python
import pickle 
import os
import matplotlib.pyplot as plt

recog_result_root_dir = './recognition_results/'

pkls = [ele for ele in os.listdir(recog_result_root_dir) if ele.endswith('.pkl')]
pkls = sorted(pkls, key=lambda x:int(x.split('.')[0]))

for pkl_name in pkls[3200:]:
    full_path = os.path.join(recog_result_root_dir, pkl_name)
    faces_list = pickle.load(open(full_path, 'rb'))
    
    print('Processing {}: faces num = {}'.format(pkl_name, len(faces_list)))

    plt.figure()
    for index, face_obj in enumerate(faces_list):
        if face_obj['face'].shape[1] == 0:
            continue
        plt.subplot(1, len(faces_list), index + 1)
        plt.imshow(face_obj['face'][:,:,::-1])
        plt.title('{}'.format(face_obj['id']))
        plt.axis('off')
    plt.show()
```
